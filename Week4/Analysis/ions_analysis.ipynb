```python
import numpy as np
import matplotlib.pyplot as plt
#load mdtraj
import mdtraj as md
import pickle
import MDAnalysis as mda
import MDAnalysis.analysis.msd as msd
```

```python
import os
# change python directory to /home/gmh4/projects/CompLab/comp-lab-class
os.chdir('/home/jh8862/theory-lab/comp-lab-2023/Week4/Analysis')
```

```python
top = '../Setup/Water-Ions/gromacs/step5_1.gro'
traj = '../Setup/Water-Ions/gromacs/md_nojump.xtc'
```

```python
topology = md.load_topology(top)
```

```python
atoms = []
```

```python
for atom in topology.atoms:
    atoms.append(atom.name)
```

```python
set(atoms)
```

```python
#compute mean squared displacement
u = mda.Universe(top, traj)
MSD = msd.EinsteinMSD(u, select='name OH2', msd_type='xyz', fft=True)
MSD.run()
```

```python
msd_sod =  MSD.results.timeseries
```

```python
msd_ow = MSD.results.timeseries
```

```python
msd_cla = MSD.results.timeseries
```

```python
nframes = MSD.n_frames
timestep = 10 # this needs to be the actual time between frames
lagtimes = np.arange(nframes)*timestep # make the lag-time axis

plt.plot(lagtimes, msd_sod, label="Sodium ions", c="m")
plt.xlabel("Time step (ps)")
plt.ylabel("MSD (A^2)")
plt.title("Mean Square Displacement of Sodium")
plt.legend()
#plt.savefig("../Figures/MSD_sodium.png", dpi=300)
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(lagtimes, msd_ow)
ax.set_xlabel("Time step (ps)")
ax.set_ylabel("MSD (A^2)")
ax.set_title("Mean Square Displacement of Water Oxygens")
plt.savefig("../Figures/MSD_wateroxygen.png", dpi=300)
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(lagtimes, msd_cla)
ax.set_xlabel("Time step (ps)")
ax.set_ylabel("MSD (A^2)")
ax.set_title("Mean Square Displacement of Chlorine Atoms")
plt.savefig("../Figures/MSD_chlorine.png", dpi=300)
plt.show()
```

```python
plt.plot(lagtimes, msd_sod, label="Sodium ions", c="m")
plt.plot(lagtimes, msd_ow, label = "Water Oxygens", c="c")
plt.plot(lagtimes, msd_cla, label = "Chlorine ions", c="y")
plt.xlabel("Time step (ps)")
plt.ylabel("MSD (A^2)")
plt.title("Mean Square Displacement of Solution Components")
plt.legend()
plt.savefig("../Figures/MSD_3on1graph.png", dpi=300)
plt.show()
```
