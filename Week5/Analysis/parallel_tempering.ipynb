```python
#Free Energy Surface
traj = md.load("../Inputs/T300/adp_exchange3temps.trr", top="../Inputs/T300/adp_exchange3temps.gro")
```

```python
phi = md.compute_phi(traj)
psi = md.compute_psi(traj)
```

```python
phi_arr = np.asarray(phi[1]).flatten()
psi_arr = np.asarray(psi[1]).flatten()
```

```python
hist, x_edges, y_edges = np.histogram2d(phi_arr, psi_arr, bins=20)
```

```python
hist_log = hist
hist_log[hist_log == 0] = 1
```

```python
F = -np.log(hist_log)
```

```python
min_F = np.min(F)
```

```python
F_normalized = F - min_F
```

```python
X, Y = np.meshgrid(x_edges, y_edges)
```

```python
plt.contourf(X[:-1, :-1], Y[:-1, :-1], F_normalized.T, cmap='viridis', vmax=10)
plt.colorbar(label='F(phi, psi) - min(F(phi, psi))')
plt.xlabel('Phi')
plt.ylabel('Psi')
plt.title('2D Histogram of Phi and Psi Data')
plt.savefig("../Figures/free_energy_300K.jpg", dpi=300, bbox_inches="tight")
plt.show()
```
