```python
water = mda.Universe("../Runs/Water/water.xyz")
water.dimensions = [12.42, 12.42, 12.42, 90, 90, 90]
```

```python
water_O = water.select_atoms("name O")
```

```python
proton = mda.Universe("../Runs/Proton/proton.xyz")
proton.dimensions = [12.42, 12.42, 12.42, 90, 90, 90]
```

```python
proton_O = proton.select_atoms("name O")
```

```python
rdf_oo = rdf.InterRDF(water_O, proton_O)
rdf_oo.run()
```

```python
plt.plot(rdf_oo.bins, rdf_oo.rdf)
plt.xlabel('Radius (A)')
plt.ylabel("g(OO)")
plt.title("RDF of Oxygens between Water and Proton traj")
plt.savefig("../Figures/rdf_water_proton.png", dpi=300, bbox_inches="tight")
plt.show()
```
