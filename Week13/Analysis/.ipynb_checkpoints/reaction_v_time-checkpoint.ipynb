```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
montime, monCV, dum, dumb, dummer, dumbest = np.loadtxt("../Outputs/monitor/MONITOR-COLVAR.metadynLog", unpack=True)
```

```python
plt.plot(montime, monCV)
plt.title("QM/MM CV over Time")
plt.xlabel("Time (fs)")
plt.ylabel("Distance between breaking C-O and making C-C bond")
plt.savefig("../Figures/CVovertime.png", dpi=300, bbox_inches="tight")
plt.show()
```
