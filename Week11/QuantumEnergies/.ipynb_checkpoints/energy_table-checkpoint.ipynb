```python
smiles_data["n_Atoms"] = smiles_data['ROMol'].map(lambda x: x.GetNumAtoms())
smiles_data.head()
```

```python
from psikit import Psikit
pk = Psikit(debug=True,threads=1,memory=12)
```

```python
optimized_scf = []
```

```python
for i, row in smiles_data.iterrows():
    pk.read_from_smiles(row["smiles"])
    %time
    print("optimizing ", row["smiles"])
    pk.energy()
    scf = pk.optimize()
    optimized_scf.append(scf)
```

```python
smiles_data["Optimized SCF"] = optimized_scf
```
