```python
#phi and psi from plumed
A = plumed.read_as_pandas("../Inputs/DialaA/COLVAR")
B = plumed.read_as_pandas("../Inputs/DialaB/COLVAR")
```

```python
#phi and psi from xtc
trajA = md.load('../Inputs/DialaA/traj_comp.xtc', top='../Inputs/DialaA/dialaA.pdb')
trajB = md.load('../Inputs/DialaB/traj_comp.xtc', top='../Inputs/DialaB/dialaB.pdb')
```

```python
phiA = md.compute_phi(trajA)
psiA = md.compute_psi(trajA)
```

```python
phiB = md.compute_phi(trajB)
psiB = md.compute_psi(trajB)
```

```python
plt.scatter(A["phi"], A["psi"], s=1, c="c", label="COLVAR A")
plt.scatter(B["phi"], B["psi"], s=1, c="m", label="COLVAR B")

plt.scatter(phiA[1].flatten(), psiA[1].flatten(), s=1, c="c", label="A from Traj")
plt.scatter(phiB[1].flatten(), psiB[1].flatten(), s=1, c="m", label="B from Traj")


plt.xlabel("phi (rad)")
plt.ylabel("psi (rad)")
plt.title("Phi v Psi from COLVAR and trajectory")
plt.legend()

plt.savefig("../Figures/exercise1.jpg", dpi=300, bbox_inches='tight')
plt.show()
```

```python
print("Std phi run_A:",np.std(A["phi"]))
print("Std phi run_B:",np.std(B["phi"]))
```

```python
fes = plumed.read_as_pandas("../Inputs/DialaA/fes.dat")
fes.head()
```

```python
hist, x_edges, y_edges = np.histogram2d(fes["phi"], fes["psi"], bins=(33,11))
X, Y = np.meshgrid(x_edges, y_edges)
```

```python
free_E = fes.pivot(index='psi', columns='phi', values='file.free')
free_E_arr = free_E.values
```

```python
hist.shape
```

```python
plt.contourf(X[:-1, :-1], Y[:-1, :-1], free_E_arr, cmap='viridis')
plt.colorbar(label='Free Energy (kJ/mol)')
plt.xlabel('Phi')
plt.ylabel('Psi')
plt.title('Free Energy Surface of Phi and Psi Metadynamics')
plt.savefig("../Figures/phipsimetadynamics_fes_contourplt.jpg", dpi=300, bbox_inches="tight")
plt.show()
```
